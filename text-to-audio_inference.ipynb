```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import os

os.listdir()
```

```python
wav_files = ['washing_machine_beginning.wav','sink1.wav','Espresso Machine.wav','water_sink.wav','Sink_on_off.wav']
```

```python
def plot_logmel_spectrogram(wav_file_path, axs, title_fontsize=10):
    y, sr = librosa.load(wav_file_path)
    mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)
    log_mel_spectrogram = librosa.power_to_db(mel_spectrogram, ref=np.max)

    title = wav_file_path.split('/')[-1]

    librosa.display.specshow(log_mel_spectrogram, sr=sr, x_axis='time', y_axis='mel', ax=axs)
    axs.set_title(title, fontsize=title_fontsize)
```

```python
nrows = 5
ncols = 1
fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=(5,15))

for i in range(nrows):
    plot_logmel_spectrogram(wav_files[i], axs[i])

plt.tight_layout()

plt.savefig('plt_rows.jpg')

from google.colab import files
files.download('plt_rows.jpg')
```

```python
# cols
nrows = 1
ncols = 5
fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=(20, 3))

for i in range(ncols):
    plot_logmel_spectrogram(wav_files[i], axs[i])

plt.tight_layout()

plt.savefig('plt_cols.jpg')

from google.colab import files
files.download('plt_cols.jpg')
```
